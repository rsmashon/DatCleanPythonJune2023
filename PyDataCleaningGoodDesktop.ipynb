# Filling different values for Null in different columns

```python
df8=df.fillna({'society': 'replaceForSociety',
              'balcony': 'replaceForBalcony'})#Fill using dictionary for each column
df8
```

# Fill with mean of the numeric column eg for bath balcony price

```python
df9=df.fillna(value=df['balcony'].mean())  # check vaaaaaalue in balcony for 13316 as 1.584376
df9
```

```python
df10=df.fillna(value=df['balcony'].min())  # fill with minimum value for the column
df10
```

# delete rows with NULL value

```python
df11=df.dropna()   #rows 2,4 13316 deleted, these had Null values
df11
```

```python
df12=df.dropna(how='any') # Deletes row if ANY  field is NULL
df12
```

```python
df12=df.dropna(how='all') # #rows 2,4 13316 NOT deleted, these had just one Null field

df12
```

```python
import numpy as np
df13=df.replace(to_replace=np.nan, value='MyChoiceValueOrNum') #replace Null value with any value
df13
```

```python
df14=df.replace(to_replace=3, value='55') #replace value   with given value, check in balcony
df14
```

# INTERPOLATE()--Linear,

```python
df['balcony']=df['balcony'].interpolate(method='linear')  #fills with mean of prceeding and next value, bacony value for 13316 is mean of 1 and 0
df
```

LIMIT direction- forward or reverse

```python
df['balcony']=df['balcony'].interpolate(method='linear',limit_direction='forward')  #fills with mean of prceeding and next value, bacony value for 13316 is mean of 1 and 0 in forward direction
df
```
